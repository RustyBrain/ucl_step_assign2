```python
import pandas as pd
```

```python
df = pd.read_csv('outputs/pollution_city_comparison.csv', index=False)
```

```python
df
```

```python
city_avg = df.groupby(['city', 'measurement', 'pollutant'], as_index=False).mean()
city_avg
```

```python
city_pivot = city_avg.pivot(index='city', columns='pollutant', values=['high', 'low', 'midnight'])
city_pivot.columns = ['_'.join(col) for col in city_pivot.columns.values]
city_pivot.to_csv('data/CSV/city_pollutants.csv')
```
