```python
# This code will retrieve weather conditions for every day within the date range specified. It requires and API key for 
# visual-crossing-weather
import requests
import pandas as pd
from io import StringIO
from API_key import key
```

```python
# Set comparator cities
locations=['Amsterdam,NL', 'Berlin,DE', 'Frankfurt,DE', 'Manchester,UK', 'London,UK', 'Madrid,ES', 'Milano,IT', 'Paris,FR']
```

```python
url = "https://visual-crossing-weather.p.rapidapi.com/history"

headers = {
    'x-rapidapi-host': "visual-crossing-weather.p.rapidapi.com",
    'x-rapidapi-key': key
    }

for location in locations:
    # This is where the date range is specified
    querystring = {"dayStartTime":"8:00:00","contentType":"csv","dayEndTime":"17:00:00","shortColumnNames":"false","startDateTime":"2016-01-01T00:00:00","aggregateHours":"24","location":location,"endDateTime":"2020-01-01T00:00:00","unitGroup":"metric"}

    response = requests.request("GET", url, headers=headers, params=querystring)

    df = pd.read_csv(StringIO(response.text))
    # Saves the weather data as a csv file in the outputs folder
    df.to_csv('outputs/' + location[:-3] + '_weather.csv', index=False)
```
